### The Code

```python
# =============================================================================
# 1. SETUP AND CONFIGURATION
# =============================================================================
import os

# Set environment variables
os.environ['HF_HOME'] = '/workspace/huggingface_cache'
os.environ['TOKENIZERS_PARALLELISM'] = 'false'

import torch
import transformers
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
    EarlyStoppingCallback,
    DataCollatorForSeq2Seq,
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer
import wandb


# Enable tf32 for better performance on Ampere GPUs
torch.backends.cuda.matmul.allow_tf32 = True


class TrainingConfig:
    """
    Configuration settings for the fine-tuning job.
    """
    # Model and dataset configuration
    MODEL_ID = "CohereLabs/aya-expanse-8b"
    DATASET_NAME = "RobbedoesHF/dutch-definitions"
    DATASET_TRAIN_SPLIT = "train"
    DATASET_EVAL_SPLIT = "validation"

    # LoRA configuration
    LORA_R = 32
    LORA_ALPHA = 64
    LORA_DROPOUT = 0.1
    # Target modules for Aya/Command models (similar to Llama)
    LORA_TARGET_MODULES = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"]

    # Training configuration
    OUTPUT_DIR = "./results_aya_expanse_8b"
    BATCH_SIZE = 4  # A good starting point for an A100 GPU
    GRAD_ACCUMULATION_STEPS = 4  # Effective batch size = 16
    LEARNING_RATE = 2e-4
    NUM_TRAIN_EPOCHS = 5
    MAX_SEQ_LENGTH = 512 
    LOGGING_STEPS = 10
    SAVE_STEPS = 100
    EVAL_STEPS = 100
    OPTIMIZER = "paged_adamw_8bit"
    LR_SCHEDULER_TYPE = "cosine"
    WARMUP_RATIO = 0.1
    EARLY_STOPPING_PATIENCE = 5
    SEED = 42

    # W&B configuration
    WANDB_PROJECT = "dutch-definition-modeling"
    WANDB_RUN_NAME = "aya-expanse-8b-qlora-dutch-definitions"

    # Test run configuration
    IS_TEST_RUN = False
    TEST_RUN_SIZE = 100

config = TrainingConfig()

# --- Weights & Biases Login ---
try:
    wandb.login()
    os.environ["WANDB_PROJECT"] = config.WANDB_PROJECT
except Exception as e:
    print(f"Could not log in to W&B: {e}. Disabling monitoring.")
    os.environ["WANDB_DISABLED"] = "true"
```

```python
# =============================================================================
# 2. MODEL AND TOKENIZER LOADING
# =============================================================================
print(f"Loading base model: {config.MODEL_ID}")

# Configure quantization for 4-bit loading (QLoRA)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16, # Use bfloat16 for A100
    bnb_4bit_use_double_quant=True,
)

# Load the Causal LM model with quantization and Flash Attention 2
model = AutoModelForCausalLM.from_pretrained(
    config.MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    attn_implementation="sdpa", # Essential for performance on A100
)

tokenizer = AutoTokenizer.from_pretrained(config.MODEL_ID)
# SFTTrainer will handle padding, but we must set a pad token
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'right' # SFTTrainer recommends right-side padding
```

```python
# =============================================================================
# 3. DATASET PREPARATION
# =============================================================================

# This function manually creates the full prompt string with special tokens.
# It's passed directly to the SFTTrainer.
def format_prompt(example):
    """
    Creates a single formatted prompt string from a single data example.
    """
    # This function now processes a single sample at a time
    system_prompt = "Je bent een expert-lexicograaf die definities schrijft voor een Nederlands woordenboek."
    user_prompt = f"Breid de volgende korte definitie voor het woord '{example['Lemma']}' uit tot een volledige definitie: '{example['DefinitionShort']}'"
    full_user_prompt = f"{system_prompt}\n\n{user_prompt}"
    
    # Manually construct the string with the exact special tokens
    text = f"<|START_OF_TURN_TOKEN|><|USER_TOKEN|>{full_user_prompt}<|END_OF_TURN_TOKEN|><|START_OF_TURN_TOKEN|><|CHATBOT_TOKEN|>{example['DefinitionFull']}<|END_OF_TURN_TOKEN|>"
    return text

def preprocess_dataset(examples):
    """
    Applies the formatting to a batch of examples and adds the result to a new 'text' column.
    """
    examples["text"] = [format_prompt(ex) for ex in [dict(zip(examples, t)) for t in zip(*examples.values())]]
    return examples
    
# --- Load and Clean the Dataset ---
dataset = load_dataset(config.DATASET_NAME)

def is_valid_example(example):
    return (
        example.get("Lemma") and isinstance(example["Lemma"], str) and len(example["Lemma"].strip()) > 0 and
        example.get("DefinitionShort") and isinstance(example["DefinitionShort"], str) and len(example["DefinitionShort"].strip()) > 0 and
        example.get("DefinitionFull") and isinstance(example["DefinitionFull"], str) and len(example["DefinitionFull"].strip()) > 0
    )

print("Filtering datasets for invalid rows...")
train_dataset = dataset[config.DATASET_TRAIN_SPLIT].filter(is_valid_example, num_proc=4)
eval_dataset = dataset[config.DATASET_EVAL_SPLIT].filter(is_valid_example, num_proc=4)

# --- Pre-process and Format the Dataset ---
print("Formatting datasets...")
train_dataset = train_dataset.map(preprocess_dataset, batched=True, num_proc=4, remove_columns=train_dataset.column_names)
eval_dataset = eval_dataset.map(preprocess_dataset, batched=True, num_proc=4, remove_columns=eval_dataset.column_names)

if config.IS_TEST_RUN:
    train_dataset = train_dataset.select(range(config.TEST_RUN_SIZE))
    eval_dataset = eval_dataset.select(range(config.TEST_RUN_SIZE))

# --- Print Final Dataset Sizes ---
print("\n" + "="*50)
print(f"Final training dataset size: {len(train_dataset)}")
print(f"Final evaluation dataset size: {len(eval_dataset)}")
print("="*50 + "\n")

# This is the reversed logic to find ONLY the invalid examples
def is_invalid_example(example):
    return not is_valid_example(example)

# Filter the original splits to get the invalid rows
invalid_train_examples = dataset['train'].filter(is_invalid_example, num_proc=4)
invalid_eval_examples = dataset['validation'].filter(is_invalid_example, num_proc=4)

# --- Print results for the training set ---
print("\n" + "="*50)
print(f"Found {len(invalid_train_examples)} invalid example(s) in the training set.")
if len(invalid_train_examples) > 0:
    print("Showing up to 5 invalid training examples:")
    # Iterate and print the first 5
    for i in range(min(5, len(invalid_train_examples))):
        print(f"  - Example {i+1}: {invalid_train_examples[i]}")

# --- Print results for the evaluation set ---
print("\n" + "="*50)
print(f"Found {len(invalid_eval_examples)} invalid example(s) in the evaluation set.")
if len(invalid_eval_examples) > 0:
    print("Showing up to 5 invalid evaluation examples:")
    # Iterate and print the first 5
    for i in range(min(5, len(invalid_eval_examples))):
        print(f"  - Example {i+1}: {invalid_eval_examples[i]}")

print("="*50 + "\n")
```

```python
# =============================================================================
# 4. SFT TRAINER SETUP
# =============================================================================
print("Configuring SFTTrainer...")

peft_config = LoraConfig(
    r=config.LORA_R,
    lora_alpha=config.LORA_ALPHA,
    lora_dropout=config.LORA_DROPOUT,
    target_modules=config.LORA_TARGET_MODULES,
    bias="none",
    task_type="CAUSAL_LM",
)

print("Setting up training arguments...")

training_arguments = TrainingArguments(
    output_dir=config.OUTPUT_DIR,
    per_device_train_batch_size=config.BATCH_SIZE,
    gradient_accumulation_steps=config.GRAD_ACCUMULATION_STEPS,
    gradient_checkpointing=True,
    optim=config.OPTIMIZER,
    learning_rate=config.LEARNING_RATE,
    num_train_epochs=config.NUM_TRAIN_EPOCHS,
    fp16=False, # bf16 is used instead
    fp16_full_eval=False,      # ← evaluate in float32
    bf16=True, # Use bfloat16 for A100 performance
    max_grad_norm=0.3,
    lr_scheduler_type=config.LR_SCHEDULER_TYPE,
    warmup_ratio=config.WARMUP_RATIO,
    logging_dir=f"{config.OUTPUT_DIR}/logs",
    eval_strategy="steps",
    logging_steps=config.LOGGING_STEPS,
    save_steps=config.SAVE_STEPS,
    eval_steps=config.EVAL_STEPS,
    save_total_limit=2,
    load_best_model_at_end=True,
    report_to="wandb",
    run_name=config.WANDB_RUN_NAME,
    seed=config.SEED,
    group_by_length=True,
)

trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    peft_config=peft_config,
    tokenizer=tokenizer,
    args=training_arguments,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=config.EARLY_STOPPING_PATIENCE)],
    dataset_text_field="text", # Tell the trainer to use our new "text" column
    # formatting_func= format_prompt, # REMOVE THIS LINE
    # This is the sequence length for the entire formatted string
    max_seq_length=config.MAX_SEQ_LENGTH,
    packing=True, # Packs multiple short examples into one sequence for efficiency
    #eval_packing=False, # Disable packing just for evaluation
)
```

```python
# =============================================================================
# 5. TRAINING
# =============================================================================

print("Starting model training with SFTTrainer...")
trainer.train()

print("\nSaving final LoRA adapter...")
final_model_path = os.path.join(config.OUTPUT_DIR, "final_checkpoint")
trainer.save_model(final_model_path)
print(f"Final model saved to {final_model_path}")

wandb.finish()
```
